```python
#FINAL
from tkinter import *


def calculate_bmi():
    kg = int(weight_tf.get())
    m = int(height_tf.get())/100
    bmi = kg/(m*m)
    bmi = round(bmi, 1)
    bmi_index(bmi)

def bmi_index(bmi):
    
    if bmi < 18.5:
        print( f'BMI = {bmi} is Underweight')
    elif (bmi > 18.5) and (bmi < 24.9):
        print( f'BMI = {bmi} is Normal')
    elif (bmi > 24.9) and (bmi < 29.9):
        print( f'BMI = {bmi} is Overweight')
    elif (bmi > 29.9):
        print( f'BMI = {bmi} is Obesity') 
    else:
        print( 'something went wrong!')   

ws = Tk()
ws.title('BMI Calculator')
ws.geometry('350x450')
ws.config(bg='black')

var = IntVar()

frame = Frame(
    ws,
    padx=10, 
    pady=10
)
frame.pack()


age = Label(frame,text="Enter Age", height=2, width=18, bg="black", fg="white" ,font=("Arial", 15)).pack(fill=X)

age = Entry(frame, width=25,bg="white",fg="blue",font=("Arial", 15)).pack(fill=X)

gen_lb = Label(
    frame,
    text='Select Gender', height=2, width=18, bg="black", fg="white" ,font=("Arial", 15)
)
gen_lb.pack(fill=X)

frame2 = Frame(
    frame
)
frame2.pack()

male_rb = Radiobutton(
    frame2,
    text = 'Male',
    variable = var,
    value = 1, width=22, bg="white", fg="blue",font=("Arial", 15)
)
male_rb.pack(fill=X)

female_rb = Radiobutton(
    frame2,
    text = 'Female',
    variable = var,
    value = 2, width=22, bg="white", fg="blue",font=("Arial", 15)
)
female_rb.pack(fill=X)

height_lb = Label(
    frame,
    text="Enter Height (cm)  ", height=2, width=18, bg="black", fg="white" ,font=("Arial", 15)
)
height_lb.pack(fill=X)

weight_lb = Label(
    frame,
    text="Enter Weight (kg)  ", height=2, width=18, bg="black", fg="white" ,font=("Arial", 15)

)
height_tf = Entry(
    frame,width=25,bg="white",fg="blue",font=("Arial", 15)
)
height_tf.pack(fill=X)
weight_lb.pack(fill=X)


weight_tf = Entry(
    frame,width=25,bg="white",fg="blue",font=("Arial", 15)
)
weight_tf.pack(fill=X)

frame3 = Frame(
    frame
)
frame3.pack()

cal_btn = Button(
    frame3,
    text='Calculate',
    command=calculate_bmi, bg="blue", fg="white",font=("Arial", 20)
)
cal_btn.pack(fill=X)



ws.mainloop()
```
